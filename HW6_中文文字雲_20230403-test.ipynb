## 參數設定

```python
url='https://news.ltn.com.tw/topic/中職' # 自由時報 #網址可更換
scroll_times = 10 # scroll down 次數
```

```python
# Chrome 參數
options = webdriver.ChromeOptions()
# 不載入圖片，加速網頁擷取
prefs = {
    'profile.default_content_setting_values' : {
        'images' : 2
    }
}
options.add_argument('headless')
options.add_experimental_option('prefs',prefs)
driver = webdriver.Chrome(options = options)
```

```python
# scroll down 10 次
driver.get(url)
for i in range(scroll_times):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(1)   
```

```python
from selenium.webdriver.common.by import By

data=''
for item in driver.find_elements(By.CSS_SELECTOR, 'a.tit'):
#     if item.text.strip() == '' or item.text.strip().startswith('三立')          or item.text.strip().startswith('聯合') or item.text.strip().startswith('中時')          or item.text.strip().startswith('自由時報') or item.text.strip().startswith('新頭殼')          or item.text.strip().startswith('公視新聞網') or item.text.strip().startswith('電腦王阿達')          or item.text.strip().startswith('udn') or item.text.strip().startswith('NOWnews')          or item.text.strip().startswith('中天') or item.text.strip().startswith('ETtoday')          or item.text.strip().startswith('Yahoo') or item.text.strip().startswith('查看完整報導'):
#         continue       
    print(item.text)
    data += item.text + '\n'
```

```python
# close chrome
driver.quit()
```

```python
# save file
with open('./news3.txt', 'w', encoding='utf8') as f:
    f.write(data)
```

## 中文文字雲

```python
from wordcloud import WordCloud, STOPWORDS
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import jieba
import jieba.analyse
from collections import Counter # 次數統計

dictfile = "./dict.txt"  # 字典檔
stopfile = "./stopwords.txt"  # stopwords
fontpath = "C:\Windows\Fonts\msjh.ttc"  # 字型檔 (注意副檔名要標,否則於電腦 C 槽內 c:\Windows\Fonts 直接尋找字型檔會錯誤;或是直接將檔案複製到 .ipynb 所屬資料夾內,直接尋找檔案即可)

mdfile = "./news3.txt"  # 文檔
pngfile = "./mask_0401.png"  # 剛才下載存的底圖

alice_mask = np.array(Image.open(pngfile))

jieba.set_dictionary(dictfile)
jieba.analyse.set_stop_words(stopfile)

text = open(mdfile,"r",encoding="utf-8").read()

tags = jieba.analyse.extract_tags(text, topK=25)

seg_list = jieba.lcut(text, cut_all=False)
dictionary = Counter(seg_list)

freq = {}
for ele in dictionary:
    if ele in tags:
        freq[ele] = dictionary[ele]
print(freq) # 計算出現的次數

wordcloud = WordCloud(background_color="white", mask=alice_mask, contour_width=3, contour_color='steelblue', font_path= fontpath).generate_from_frequencies(freq)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
