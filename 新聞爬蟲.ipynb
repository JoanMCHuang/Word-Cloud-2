## [download chrome driver](https://sites.google.com/chromium.org/driver/)

```python
from selenium import webdriver
import sys
import time
```

## 參數設定

```python
url='https://news.ltn.com.tw/topic/台灣' # 自由時報
scroll_times = 10 # scroll down 次數
```

```python
# Chrome 參數
options = webdriver.ChromeOptions()
# 不載入圖片，加速網頁擷取
prefs = {
    'profile.default_content_setting_values' : {
        'images' : 2
    }
}
options.add_argument('headless')
options.add_experimental_option('prefs',prefs)
driver = webdriver.Chrome(options = options)
```

```python
# scroll down 10 次
driver.get(url)
for i in range(scroll_times):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(1)   
```

```python
from selenium.webdriver.common.by import By

data=''
for item in driver.find_elements(By.CSS_SELECTOR, 'a.tit'):
#     if item.text.strip() == '' or item.text.strip().startswith('三立')          or item.text.strip().startswith('聯合') or item.text.strip().startswith('中時')          or item.text.strip().startswith('自由時報') or item.text.strip().startswith('新頭殼')          or item.text.strip().startswith('公視新聞網') or item.text.strip().startswith('電腦王阿達')          or item.text.strip().startswith('udn') or item.text.strip().startswith('NOWnews')          or item.text.strip().startswith('中天') or item.text.strip().startswith('ETtoday')          or item.text.strip().startswith('Yahoo') or item.text.strip().startswith('查看完整報導'):
#         continue       
    print(item.text)
    data += item.text + '\n'
```

```python
# close chrome
driver.quit()
```

```python
# save file
with open('./news.txt', 'w', encoding='utf8') as f:
    f.write(data)
```
